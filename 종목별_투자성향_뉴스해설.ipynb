```python
df = pd.read_csv('뉴스데이터_최종_0727.csv')
df.head()
```

```python
df.info()
```

```python
# list로 변환

# 문자열에서 '[', ']', 공백, 따옴표를 제거하고 쉼표로 분할하여 리스트로 변환
def parse_keywords(keywords):
    # 문자열에서 '[', ']' 및 공백을 제거
    keywords = keywords.strip('[]').replace("'", "").replace(' ', '')
    # 쉼표로 분할하여 리스트로 변환
    return keywords.split(',')

# 데이터프레임의 'keywords' 열에 적용
df['keywords'] = df['keywords'].apply(parse_keywords)
df['keywords'][0][0]
```

```python
df.keywords
```

```python
# 모든 키워드를 중복 없이 집합으로 통합
all_keywords = set(keyword for sublist in df['keywords'] for keyword in sublist)

# 각 키워드가 속한 인덱스를 저장할 딕셔너리
keyword_indices = {keyword: [] for keyword in all_keywords}

# 각 키워드에 대해 해당 인덱스를 저장
for idx, keywords in df['keywords'].items():
    for keyword in keywords:
        keyword_indices[keyword].append(idx)

# 결과를 데이터프레임으로 변환
result_df = pd.DataFrame(list(keyword_indices.items()), columns=['keyword', 'index_set'])
result_df
```

```python
df_type = pd.read_csv('InvestmentPropensity/종목분류지표_최종.csv', encoding='cp949')
df_type
```

```python
# 데이터프레임 병합
merged_df = pd.merge(df_type, result_df, left_on='기업명', right_on='keyword', how='left')
merged_df = merged_df.drop(columns=['keyword'])
merged_df.dropna(inplace=True)
merged_df
```

```python
import random
# 뉴스가 여러개 존재할 경우 랜덤으로 하나만 select
def get_random_element(lst):
    return str(random.choice(lst)) if lst else ''

merged_df['index_num'] = merged_df['index_set'].apply(get_random_element)
merged_df
```

```python
merged_df['news_date'] = np.nan

for index, row in merged_df.iterrows():
    # index_num의 첫 번째 요소를 가져옵니다 (여기서는 하나의 요소만 있습니다)
    idx_str = row['index_num']
    
    # 문자열을 정수로 변환
    try:
        idx = int(idx_str)
        
        # df에서 해당 인덱스의 news_dates를 가져옵니다
        if 0 <= idx < len(df):
            merged_df.at[index, 'news_date'] = df.loc[idx, 'news_dates']
            merged_df.at[index, 'news_summary'] = df.loc[idx, 'summary']
    except ValueError:
        # index_num이 정수로 변환할 수 없는 경우
        merged_df.at[index, 'news_date'] = np.nan
        merged_df.at[index, 'news_summary'] = np.nan
merged_df
```
