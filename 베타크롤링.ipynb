# 베타 크롤링하기

```python
import requests
from bs4 import BeautifulSoup

def get_beta_from_yahoo_finance(ticker):
    url = f"https://finance.yahoo.com/quote/{ticker}.KS"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # HTML 구조가 다를 수 있으므로, 직접 확인한 후 적절한 태그 사용
    summary_table = soup.find('div', {'id': 'quote-summary'})
    if summary_table:
        beta_row = summary_table.find('td', text='Beta (5Y Monthly)')
        if beta_row:
            beta_value = beta_row.find_next('td').text
            return beta_value
    return 'N/A'

ticker = '138930'  # 삼성전자의 티커 심볼 예제
beta = get_beta_from_yahoo_finance(ticker)
print(f"{ticker}.KS의 5년 월별 베타 값: {beta}")
```

```python
import requests
import pandas as pd
from datetime import datetime
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor

def fetch_page_data(url):
    html = requests.get(url, headers={'User-agent': 'Mozilla/5.0'}).text
    return pd.read_html(html, header=0)[0]

def get_stock_data(stock_code, start_date, end_date):
    # 시작 및 종료 날짜를 datetime 형식으로 변환
    start_date = datetime.strptime(start_date, '%Y-%m-%d')
    end_date = datetime.strptime(end_date, '%Y-%m-%d')
    
    # 첫 페이지 URL
    url = f'https://finance.naver.com/item/sise_day.nhn?code={stock_code}&page=1'
    html = requests.get(url, headers={'User-agent': 'Mozilla/5.0'}).text
    
    # BeautifulSoup으로 HTML 파싱
    bs = BeautifulSoup(html, 'lxml')
    
    # 마지막 페이지 번호 찾기
    pgrr = bs.find('td', class_='pgRR')
    last_page = int(pgrr.a['href'].split('=')[-1])
    
    # 빈 데이터프레임 생성
    df = pd.DataFrame()
    sise_url = f'https://finance.naver.com/item/sise_day.nhn?code={stock_code}'
    
    # 병렬로 데이터를 가져오기 위해 ThreadPoolExecutor 사용
    with ThreadPoolExecutor(max_workers=8) as executor:
        futures = []
        for page in range(1, last_page + 1):
            url = f'{sise_url}&page={page}'
            futures.append(executor.submit(fetch_page_data, url))
        
        for future in futures:
            df = df.append(future.result(), ignore_index=True)
    
    # NaN 값 제거
    df = df.dropna()
    
    # 날짜를 datetime 형식으로 변환
    df['날짜'] = pd.to_datetime(df['날짜'], format='%Y.%m.%d')
    
    # 시작 날짜와 끝 날짜로 데이터 필터링
    df = df[(df['날짜'] >= start_date) & (df['날짜'] <= end_date)]
    
    # 인덱스 재설정
    df = df.reset_index(drop=True)
    
    return df

def calculate_daily_return_std(df):
    # 일별 수익률 계산
    df['수익률'] = df['종가'].pct_change()
    # 일별 수익률의 표준편차 계산
    daily_return_std = df['수익률']
    return daily_return_std

# def get_all_stocks_daily_return(stock_codes, start_date, end_date):
#     results = []
#     for stock_code in stock_codes:
#         try:
#             stock_data = get_stock_data(stock_code, start_date, end_date)
#             daily_std = calculate_daily_return_std(stock_data)
#             results.append({'종목 코드': stock_code, '일별 수익률': daily_std})
#         except Exception as e:
#             results.append({'종목 코드': stock_code, '일별 수익률': None})
#             print(f"Error processing stock code {stock_code}: {e}")
#     return pd.DataFrame(results)
```
